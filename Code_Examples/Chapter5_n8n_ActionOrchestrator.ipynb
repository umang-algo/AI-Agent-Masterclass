# Interactive Agent Session: Chapter 5 — n8n Intelligent VIP Escalation Swarm

We have reached the **Final Tier** of agent development: **Real-World Impact.**

**The Business Problem:** A VIP CEO of a major account is reporting a critical system failure that is costing them revenue. A simple chat reply is not enough. We need an **Autonomous Action Layer** to mobilize the technical team.

**Our Agent's Job:**
1. Detect the high-stakes nature of the request.
2. Trigger an **n8n Webhook** that coordinates a multi-app response (Slack Alert + CRM Update + Priority Support Ticket).

**Why n8n?** Because coding 500+ integrations (Slack, Gmail, HubSpot) manually is inefficient. n8n acts as the workflow glue for your agents.

**Prerequisites:** You can use a local n8n instance or an n8n Cloud Webhook URL.

```python
import os
import requests
import json
from datetime import datetime
from dotenv import load_dotenv
from IPython.display import display, HTML

load_dotenv()

# ── 0. Configuration ──────────────────────────────────────────────────────────
N8N_WEBHOOK_URL = os.getenv("N8N_WEBHOOK_URL") or "https://your-n8n-instance.com/webhook/vip-esc-99"

if not os.getenv("N8N_WEBHOOK_URL"):
    USE_SIMULATION = True
    print("⚠️  No n8n Webhook found — running in high-fidelity simulation mode.")
else:
    USE_SIMULATION = False
    print("✅ n8n Action Layer Initialized.")

# ── 1. The VIP Escalation Trigger ───────────────────────────────────────────
def trigger_vip_recovery_swarm(customer_name: str, email: str, fault_type: str, severity: int):
    # This payload is designed to be parsed by an n8n 'Webhook' node
    payload = {
        "event": "CRITICAL_VIP_FAULT",
        "meta": {
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "source": "AI_Compliance_Auditor_Node",
            "priority": "P0_EMERGENCY" if severity > 8 else "P1_URGENT"
        },
        "customer_data": {
            "name": customer_name,
            "email": email
        },
        "incident": {
            "type": fault_type,
            "description": f"Automated detection of {fault_type} for VIP account."
        }
    }
    
    if USE_SIMULATION:
        return {"status": "success", "flow": "Payment_Recovery_Workflow", "payload": payload}
    
    try:
        # Syncing code logic to trigger n8n
        response = requests.post(N8N_WEBHOOK_URL, json=payload, timeout=10)
        response.raise_for_status()
        return {"status": "success", "flow": "Live_n8n_Production", "payload": payload}
    except Exception as e:
        print(f"❌ Webhook Dispatch Failed: {e}")
        return {"status": "simulation", "payload": payload}

# ── 2. Premium "Action Signal" UI ───────────────────────────────────────────
def render_n8n_signal_ui(data: dict):
    payload_json = json.dumps(data['payload'], indent=2)
    
    html = f'''
    <style>
        @import url("https://fonts.googleapis.com/css2?family=Unbounded:wght@800&family=Outfit:wght@300;400;700&display=swap");
        .sig-container {{ max-width: 850px; margin: 30px auto; font-family: "Outfit", sans-serif; background: #000; padding: 45px; border-radius: 40px; border: 1px solid #222; box-shadow: 0 50px 100px rgba(0,0,0,1); color: #fff; overflow: hidden; }}
        .sig-header {{ display: flex; align-items: center; justify-content: space-between; margin-bottom: 40px; border-left: 6px solid #f43f5e; padding-left: 20px; }}
        .sig-title {{ font-family: "Unbounded", sans-serif; font-size: 20px; font-weight: 800; background: linear-gradient(90deg, #f43f5e, #ec4899); -webkit-background-clip: text; -webkit-text-fill-color: transparent; }}
        .sig-meta {{ background: #111; padding: 25px; border-radius: 20px; border: 1px solid #333; margin-top: 25px; }}
        .sig-label {{ font-size: 10px; text-transform: uppercase; letter-spacing: 2px; color: #f43f5e; font-weight: 800; margin-bottom: 10px; display: block; }}
        .sig-code {{ font-family: monospace; color: #a5f3fc; font-size: 11px; white-space: pre-wrap; line-height: 1.5; }}
        .sig-success {{ margin-top: 25px; background: #064e3b; color: #6ee7b7; padding: 15px; border-radius: 12px; font-size: 13px; font-weight: 700; border: 1px solid #065f46; }}
    </style>
    <div class="sig-container">
        <div class="sig-header">
            <div class="sig-title">n8n ACTION SIGNAL DISPATCHED</div>
            <div style="font-size: 11px; background: #ffffff1a; padding: 4px 12px; border-radius: 50px;">VIP_RECOVERY_PROTO</div>
        </div>
        
        <span class="sig-label">Outbound JSON Payload</span>
        <div class="sig-code">{payload_json}</div>

        <div class="sig-success">
            ✅ SIGNAL BROADCASTED: n8n workflow [<b>{data['flow']}</b>] successfully triggered for escalation.
        </div>
        <div style="margin-top: 20px; font-size: 10px; color: #444; text-align: center;">Course Proof-of-Concept • Chapter 5 • Final Step</div>
    </div>
    '''
    display(HTML(html))

# ── 3. High-Priority Execution ───────────────────────────────────────────────
RESPONSE = trigger_vip_recovery_swarm(
    customer_name="Sarah Jenkins (CEO @ CloudScale)",
    email="sjenkins@cloudscale.io",
    fault_type="CRITICAL_BILLING_FAILURE",
    severity=10
)

render_n8n_signal_ui(RESPONSE)
```
