# Interactive Agent Session: Chapter5_Dify_API

This Jupyter notebook provides a sandbox to run AI Agent code securely block-by-block. Make sure you have your `.env` configured properly before executing the agent loop below.

```python
import requests

DIFY_API_KEY = "app-your_dify_api_key_here"
DIFY_ENDPOINT = "https://api.dify.ai/v1/chat-messages"

def trigger_dify_lead_qualifier(user_message):
    headers = {
        "Authorization": f"Bearer {DIFY_API_KEY}",
        "Content-Type": "application/json"
    }
    
    payload = {
        "inputs": {}, # Any pre-defined drag-and-drop variables the PM set up in the UI
        "query": user_message, # The customer's message
        "response_mode": "blocking", # Wait for the full answer
        "conversation_id": "", 
        "user": "website-visitor-001"
    }
    
    response = requests.post(DIFY_ENDPOINT, headers=headers, json=payload)
    return response.json().get('answer')

# Simulate a website visitor
visitor_input = "Hi, I have a team of 45 people and we're looking for enterprise pricing."
print(f"Lead Agent Reply: {trigger_dify_lead_qualifier(visitor_input)}")
```
