# Interactive Agent Session: Chapter3_GoogleADK_Enterprise

This Jupyter notebook provides a sandbox to run AI Agent code securely block-by-block. Make sure you have your `.env` configured properly before executing the agent loop below.

```python
import vertexai
from vertexai.preview.generative_models import GenerativeModel, Tool
import vertexai.preview.generative_models as generative_models

# Initialize enterprise project
vertexai.init(project="my-gcp-enterprise-project", location="us-central1")

# Enable Google Search Grounding to forcefully prevent hallucinations
search_tool = Tool.from_google_search_retrieval(
    google_search_retrieval=generative_models.grounding.GoogleSearchRetrieval()
)

# Initialize the Gemini model natively packed with the search tool
model = GenerativeModel("gemini-1.5-pro-preview-0409", tools=[search_tool])

prompt = """
Read this bug report: "Prod cluster hitting intermittent Memory OOM exceptions."
Action: Use search to find the latest known public issues regarding websocket memory leaks in NGINX ingress controllers, and propose a mitigation.
"""

# Execute the chain
response = model.generate_content(prompt)
print(response.text)

# Inspect Grounding Metadata to audit the exact sources used!
for chunk in response.candidates[0].grounding_metadata.search_queries:
    print(f"Researched Query: {chunk}")
```
