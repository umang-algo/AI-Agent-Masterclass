# Interactive Agent Session: Chapter4_SmolAgents_Math

This Jupyter notebook provides a sandbox to run AI Agent code securely block-by-block. Make sure you have your `.env` configured properly before executing the agent loop below.

```python
from smolagents import CodeAgent, HfApiModel

# Initialize a completely free open-source model endpoint from the Hugging Face hub
model = HfApiModel(model_id="mistralai/Mixtral-8x7B-Instruct-v0.1")

# Create an Agent natively configured to write Code
agent = CodeAgent(tools=[], model=model, add_base_tools=True)

# Execute Prompt
# The agent will write python code internally, execute it, and respond with the exact value
response = agent.run("If a logistics company ships 12,045 packages a day at an average cost of $4.32, and inflation adds a compounding 2.5% cost per month for 12 months, what is the final daily operating cost at the end of the year?")

print(f"Mathematical Answer: {response}")
```
